```python
import threading  #módulo para a construção de multithreads
import time  #módulo para a medição de tempo
from random import randint  #módulo para geração de números randomicos
```

```python
#função que retira um número do contador

def funcao_1(num):
    n=num
    while n>0:
        n-=1  #decrementa o contador
        print("n_1: {}".format(n))   #print do contador
        time.sleep(randint(0,2))  # coloca a thread para dormir X  segundos (randint gera um vao aleatóreo entre 0 e 2)
```

```python
#função que adiciona um número para o contador

def funcao_2(num):
    n=num
    while n<100:
        n+=1  #incrementa o contador
        print("n_2: {}".format(n))   #print do contador
        time.sleep(randint(0,2))  # coloca a thread para dormir X  segundos (randint gera um vao aleatóreo entre 0 e 2)
```

```python
#colocando as threads para funcionar:

if __name__ == "__main__":

    t0 = time.time()   # tempo antes da execução  
    
    t1 = threading.Thread(target=funcao_1, args=(100,))  #instancia um objeto da classe Thread para executar a função 1
    t2 = threading.Thread(target=funcao_2, args=(0,))  #instancia um objeto da classe Thread para executar a função 2
    
    t1.start()  #inicia a Thread
    t2.start()
    
    t1.join()  #garante o fim da execução
    print("Funcao 1 terminou")
    t2.join()
    print("Funcao 2 terminou")
    
    t1 = time.time()   # tempo no fim da execução  
    
    print("Fim!")
    print("Tempo total de execução {}".format(t1-t0))
```
